```python
import pandas as pd

data = {
    'Stage': ['Visitors', 'Leads', 'Opportunities', 'Customers'],
    'Users': [1200, 500, 220, 90]
}

df = pd.DataFrame(data)
df
```

```python
import matplotlib.pyplot as plt

plt.figure()
plt.bar(df['Stage'], df['Users'])
plt.title("Marketing Funnel Analysis")
plt.xlabel("Funnel Stages")
plt.ylabel("Number of Users")
plt.show()
```
